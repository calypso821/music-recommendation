## Primerjanje klasifikacijskih tocnosti na razlicnih podatkih

```python
# Import libraries and classes required for this example:
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd 
```

```python
# Tracks_classification_1.csv - removed outliers (duration)
# Tracks_classification_2.csv - NO removed outliers!

df_track1 = pd.read_csv('Data\Genre\Tracks_classification_raw.csv', index_col=0)
df_track2 = pd.read_csv('Data\Genre\Tracks_classification_5.csv', index_col=0)
```

```python
df1 = df_track1.drop(['name', 'artist_id', 'album_id', 'key', 'mode', 'time_signature'], axis=1)
df2 = df_track2.drop(['name', 'artist_id', 'album_id'], axis=1)
```

```python
data = [df1, df2]
```

```python
for d_set in data:
    print(d_set.shape[0])
```

```python
def compare_sets(data):
    
    scores = []
    
    for d_set in data:
        # Assign values to the X and y variables
        Y = d_set.pop("genre").values
        X = d_set.values

        # Split dataset into random train and test subsets:
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)

        # Decision Tree classifier
        classifier1 = DecisionTreeClassifier()
        classifier1.fit(X_train, Y_train) 

        # Predict y data with classifier: 
        Y_predict = classifier1.predict(X_test)

        dtree = precision_score(Y_test, Y_predict, average='weighted')

        # Random forset classifier
        classifier2 = RandomForestClassifier(n_estimators=100)
        classifier2.fit(X_train, Y_train) 

        # Predict y data with classifier: 
        Y_predict = classifier2.predict(X_test)

        rforest = precision_score(Y_test, Y_predict, average='weighted')

        scores.append([dtree, rforest])

    return scores
```

```python
scores = compare_sets(data)
```

```python
print("Data set, Decision tree score, Random forest score")
for index, d_set in enumerate(scores):
    print("Set %d: %.2f %.2f" % (index, d_set[0], d_set[1]))
```
